```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gorilla = cv2.imread('Computer Vision With Python/DATA/gorilla.jpg', 0)
color_gorilla = cv2.imread('Computer Vision With Python/DATA/gorilla.jpg')
show_gorilla = cv2.cvtColor(color_gorilla, cv2.COLOR_BGR2RGB)
```

```python
def show_hist(img, title, mask=None):
    color = ('b', 'g', 'r')
    for i, col in enumerate(color):
        histr = cv2.calcHist([img], [i], mask, [256], [0,256])
        plt.plot(histr, color=col)
        plt.xlim([0,256])

    plt.title(title)
```

```python
plt.imshow(gorilla, cmap='gray')
```

```python
hist_values = cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0,256])
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
plt.imshow(eq_gorilla, cmap='gray')
```

```python
hist_values = cv2.calcHist([eq_gorilla], [0], None, [256], [0, 256])
plt.plot(hist_values)
```
