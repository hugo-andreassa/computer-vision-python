```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(384, 0), pt2=(510, 250), color=(0,255,0), thickness=10)
plt.imshow(blank_img)
```

```python
# Pt1: X 156 Y 356
# Pt2: X 356 Y 156
cv2.rectangle(blank_img, pt1=(156,356), pt2=(356, 156), color=(0,255,0), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center=(256,256), radius=250, color=(255,0,0), thickness=8)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0,512), pt2=(512,0), color=(102,255,255), thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(
    blank_img, 
    text="Hello", 
    org=(10, 500), 
    fontFace=font, 
    fontScale=4, 
    color=(255,255,255), 
    thickness=3, 
    lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int32)
```

```python
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```
