```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
horse = cv2.imread('Computer Vision With Python/DATA/horse.jpg')
show_horse = cv2.cvtColor(horse, cv2.COLOR_BGR2RGB)

img = cv2.imread('Computer Vision With Python/DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('Computer Vision With Python/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks, cv2.COLOR_BGR2RGB)
```

```python
def show_hist(img, title, mask=None):
    color = ('b', 'g', 'r')
    for i, col in enumerate(color):
        histr = cv2.calcHist([img], [i], mask, [256], [0,256])
        plt.plot(histr, color=col)
        plt.xlim([0,256])

    plt.title(title)
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
```

```python
mask[300:400, 100:400] = 255
plt.imshow(mask, cmap='gray')
```

```python
masked_img = cv2.bitwise_and(img, img, mask=mask)
show_masked_img = cv2.bitwise_and(show_rainbow, show_rainbow, mask=mask)
plt.imshow(show_masked_img)
```

```python
show_hist(img, 'RAINBOW WITH MASK', mask)
```

```python
show_hist(img, 'RAINBOW WITHOUT MASK')
```
