```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread("Computer Vision With Python/DATA/00-puppy.jpg")
type(img)
```

```python
img.shape
```

```python
# MATPLTLIB ---> RGB
# OPENCV ---> BGR
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread("Computer Vision With Python/DATA/00-puppy.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='magma')
```

```python
new_img = cv2.resize(fix_img, (400, 400))
plt.imshow(new_img)
```

```python
w_ratio = 0.2
h_ratio = 0.2
new_img = cv2.resize(fix_img, (0, 0), fix_img, w_ratio, h_ratio)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
cv2.imwrite("New_img.jpg", fix_img)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```
